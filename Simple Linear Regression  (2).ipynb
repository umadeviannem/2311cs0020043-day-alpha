```python
# Build regression model
import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily", data = data1).fit()
```

```python
model1.summary()
```

#### Observations###
- Predicted equn is y_hat = Beta_0+Beta_1 x
-  where beta_0,beta_1 = 13.8356 , 1.3397 x
-  probability for checking the significant of Beta_0,Beta_1,
-  If P<=0.05
-  the coefficient significant and useful
-  If p > 0.05
-  The coefficient significant

- The probability(p-value) for intercept(beta_0) is 0.707 > 0.05
- Therefore the intercept coefficient may not be that much significant in prediction
- However the p-value for "daily" (beta_1) is 0.00 < 0.05
- Therefore the beta_1 coefficient is highly significant and is contrinutint to prediction 

```python
# plot the linear Regression line using seaborn regplot() method
sns.regplot(x="daily", y="sunday", data=data1)
plt.xlim([0,1250])
plt.show()
```

```python
model1.params
```

```python
#print the model  statistics (t and p-Values)
print(f'model t-values:\n{model1.tvalues}\n--------------\n{model1.pvalues}')
```

Predict for new data point

```python
#print the quality of fitted line(R squared values)
model1.rsquared,model1.rsquared_adj
```

```python
# Predict for 200 and 300 daily circulation
newdata=pd.Series([200,300,1500])
```

```python
data_pred=pd.DataFrame(newdata,columns=['daily'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
# predict on all givem traing data

pred = model1.predict(data1["daily"])
pred
```

```python
# Add predicted values as a column in data1\
data1["Y_hat"] = pred
data1
```

```python
# Compute the error values (residuals) and add as another column
data1["residuals"]= data1["sunday"]-data1["Y_hat"]
data1
```

```python
# Compute Mean Squared Error for the model

mse = np.mean((data1["daily"]-data1["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```

Checking the model residuals scannter plot

```python
# Plot the residuals versus y_hat
plt.scatter(data1["Y_hat"], data1["residuals"])
```

### Observations ####

- There appears to be no trend and the residuals are randomly placed around the zero error line
- Hence the assumption of homoscedasticty is satisfied(constant varience in residuals)

```python
# P,ot the  Q-Q plot (to check the normality of residuals)
import statsmodels.api as sm
sm.qqplot(data1["residuals"], line='45', fit=True)
plt.show()
```
